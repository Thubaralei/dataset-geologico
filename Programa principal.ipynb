```python
df = pd.read_csv ('original.csv', sep = ';')
df.head()
```

```python
plt.style.use("ggplot")
plt.figure(figsize = (20, 12))
df["ESPORO"].hist(bins = 5, ec = "k", alpha = .4, color = "royalblue")
plt.title("Distribuição das dos dados")
plt.xlabel("Espessura")
plt.ylabel("MOA")
```

```python
plt.figure(figsize = (20, 12))
df[df["MOA"] < 350]["MOA"].hist(bins = 20, ec = "k", alpha = .6, color = "royalblue")
plt.title("Distribuição de acima de 900")
plt.xlabel("Espessura")
plt.ylabel("N�vel")
```
